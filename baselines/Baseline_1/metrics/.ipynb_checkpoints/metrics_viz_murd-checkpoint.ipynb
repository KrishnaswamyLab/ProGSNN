## Multi-head training

VAE: trains for coordinate reconstruction (BCE loss) and K-L divergence of latent space (from bivariate normal)
MLP: trains for regression of timestep (MSE loss) from input = VAE's latent space output (x1, x2)

Things to try:

```python
MULTIHEAD_MODEL = "vae-mlp_murd_2024-05-10-23-15"
SMOOTH_WINDOW = 20
GRID_STEP_X = 50
BURNIN_N_EPOCHS = 100

with open(f"{ROOT}/{MULTIHEAD_MODEL}/{MULTIHEAD_MODEL}_history.dat", "rb") as f:
    history_dict = pickle.load(f)
history_df = pd.DataFrame.from_records(history_dict)
del history_dict
history_df.head(10)
```

```python
print(f'best epoch saved: {history_df.best_epoch.max()}')
min_valid_loss_i = history_df.loss_valid.argmin()
min_valid_loss_epoch = history_df.epoch.iloc[min_valid_loss_i]
print(f'epoch of min valid loss:\n{history_df.loc[min_valid_loss_i]}\n')

best_valid_mse_i = history_df.mlp_mse_valid.argmin()
best_valid_mse_epoch = history_df.epoch.iloc[best_valid_mse_i]
print(f'best MLP validation MSE(timestep): {history_df.mlp_mse_valid.min():.6f} at epoch {best_valid_mse_epoch}')
print(f'best VAE validation PCC(coord dists): {history_df.mean_pcc_valid.max():.6f}')
```

```python
history_df.iloc[best_valid_mse_i]
```

```python
mf.show_train_plot(history_df=history_df,
                   title='VAE training losses',
                   metrics_l=['vae_reconst_info_loss_train', 
                              'vae_KLD_loss_scaled_train'],
                   burnin_n_epochs=BURNIN_N_EPOCHS,
                   y_lim=None,
                   y_scale_log=False,
                   smooth=True,
                   smooth_window=SMOOTH_WINDOW,
                   grid_step_x=GRID_STEP_X,
                   legend_loc=None,
                   vline_x=min_valid_loss_epoch)
```

```python
mf.show_train_plot(history_df=history_df,
                   title='MLP losses',
                   metrics_l=['mlp_loss_train', 
                              'mlp_loss_valid'],
                   burnin_n_epochs=BURNIN_N_EPOCHS,
                   y_lim=None,
                   y_scale_log=False,
                   smooth=True,
                   smooth_window=SMOOTH_WINDOW,
                   grid_step_x=GRID_STEP_X,
                   legend_loc=None,
                   vline_x=best_valid_mse_epoch)
```

```python
mf.show_train_plot(history_df=history_df,
                   title='Mean Pearson correlation: valid set inputs\' vs.'
                   '\nVAE-reconstructions\' pairwise coord. distances',
                   metrics_l=['mean_pcc_valid'],
                   smooth=True,
                   smooth_window=SMOOTH_WINDOW,
                   grid_step_x=GRID_STEP_X,
                   legend_loc=None)
```
