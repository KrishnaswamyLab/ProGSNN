## MSP mutated complexes descriptive stats

```python
# get number of atoms and residues in each 'mutated_atoms' df

n = len(msp_data)
atom_residue_cts_dictl = [None] * n
for i in range(n):
    dataset_dict = {}
    df = msp_data[i]['mutated_atoms']
    dataset_dict['n_atoms'] = len(df)
    dataset_dict['chain_residue_cts'] = {}
    total_residue_ct = 0
    for chain in df['chain'].unique():
        unique_residues_series = df[df['chain'].values == chain]['residue'].unique()
        unique_residue_ct = len(unique_residues_series)
        dataset_dict['chain_residue_cts'][chain] = unique_residue_ct
        total_residue_ct += unique_residue_ct
    dataset_dict['chain_residue_cts']['total'] = total_residue_ct
    atom_residue_cts_dictl[i] = dataset_dict
```

Get center of mass arrays for a set of mutated protein complexes (same base complex)

```python
reload(atom3d_data)

complex_idx = range(0, 10)
arrays_l = [None] * len(complex_idx)
for i, idx in enumerate(complex_idx):
    msp_df = msp_data[i]['mutated_atoms']
    resid_ctr_mass_array = atom3d_data.get_residues_ctr_mass(msp_df)
    arrays_l[i] = resid_ctr_mass_array
    print('array shape:', resid_ctr_mass_array.shape)
```

Train min-max scaler to rescale xyz (center of mass) values

```python
"""
NOTE: train min-max scaler on training set only!
"""
train_arrays_l = arrays_l

train_arrays_stack = np.row_stack(train_arrays_l)
train_mins = np.apply_along_axis(func1d=np.min, arr=train_arrays_stack, axis=0)
train_maxs = np.apply_along_axis(func1d=np.max, arr=train_arrays_stack, axis=0)
print(train_mins)
print(train_maxs)
```

Rescale each mutated protein complex's xyz (center of mass) array in the `arrays_l`

```python
ranges = train_maxs - train_mins
scale_xyz_row = lambda row: (row - train_mins) / ranges

scaled_arrays_l = [None] * len(arrays_l)
for i, arr in enumerate(arrays_l):
    arr_scaled = np.apply_along_axis(func1d=scale_xyz_row, arr=arr, axis=1)
    scaled_arrays_l[i] = arr_scaled

scaled_arrays_l[0]
```

### Shifting atoms by new residue centers

```python
# original pdb frame
frame = traj[0]
residue_ctr_coords = [None] * frame.n_residues
for j, residue in enumerate(frame.top.residues):
    atom_indices = [atom.index for atom in residue.atoms]
    # note that frame.xyz[0].shape = (n_atoms, 3)
    atom_coords = frame.xyz[0][atom_indices] 
    mean_coords = np.mean(atom_coords, axis=0)
    residue_ctr_coords[j] = mean_coords
residue_ctr_coords = np.row_stack(residue_ctr_coords)
```

```python
# model predicted residue-center coords: need array of shape (n_res, 3)
pred_resid_coords = residue_ctr_coords + np.array([1.0, 2.0, 3.0])
ctr_diff = pred_resid_coords - residue_ctr_coords
# print(ctr_diff)
ctr_diff.shape
```

```python
frame.xyz[0]
```

```python
pred_residue_ctr_coords = [None] * frame.n_residues
for j, residue in enumerate(frame.top.residues):
    print(f'residue {j}')
    atom_indices = [atom.index for atom in residue.atoms]
    print(atom_indices)
    # note that frame.xyz[0].shape = (n_atoms, 3)
    atom_coords = frame.xyz[0][atom_indices]
    print(atom_coords)

    shift_atom_coords = atom_coords + ctr_diff[j]
    print(shift_atom_coords, '\n')
    pred_residue_ctr_coords[j] = shift_atom_coords
    
pred_residue_ctr_coords = np.row_stack(pred_residue_ctr_coords)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
np.set_printoptions(precision=4)

def eucl_dist_pcc(coords_1, coords_2, squared=False):
    """
    For two sets of coordinates, calculate the Pearson correlation 
    coefficient of their two intra-set euclidean distances (each sized
    n-choose-2), optionally squared.
    """
    from sklearn.metrics.pairwise import euclidean_distances
    dists = [None] * 2
    for i, coords in enumerate((coords_1, coords_2)):
        dist = euclidean_distances(coords, coords)
        # print(dist)
        # exclude self-distances and duplicates (i,j-th = j,i-th distance)
        # create lower triangular mask excluding diagonal
        tril_mask = np.tril_indices_from(dist, k=-1)
        dist = dist[tril_mask]
        if squared:
            dist = np.square(dist)
        # print(dist)
        # print(dist.shape, '\n')
        dists[i] = dist
    # calc the 2x2 corr matrix, and return the top off-diagonal
    pcc = np.corrcoef(dists[0], dists[1])[0, 1]
    return pcc



a1 = np.random.uniform(low=-1.0, high=1.0, size=(10, 3))
# print(a1)
a2 = np.random.uniform(low=-1.0, high=1.0, size=(10, 3))
# print(a2)

# dists = euclidean_distances(a1, a2) # .flatten()
# print(dists.shape)
eucl_dist_pcc(a1, a2, False)
```
