## Mean/random models {-}

For each property, 'train' a model that always guesses the mean value of the training set. Calculate the MSE that this naive model achieves versus the true values in the training set.

```python
for property in PROPERTIES:
    train_targets = np.array(
        [x['target'][property] for x in atlas_datasets['train']]
    )
    test_targets = np.array(
        [x['target'][property] for x in atlas_datasets['test']]
    )
    # mean model preds are the mean of the train targets,
    # with as many guesses as samples in the test set
    preds = np.mean(train_targets).repeat(len(test_targets))

    # calc MSE
    mean_squared_error = MeanSquaredError()
    mean_model_mse = mean_squared_error(torch.tensor(preds), 
                                        torch.tensor(test_targets).squeeze()) \
        .detach().numpy().item()
    print(f"MSE({property}): {mean_model_mse:.6f}")
```

Random guessing of timestep: simulate the expected MSE of *n* random timesteps (scaled to (0,1)) vs. scaled timesteps.

```python
from numpy.random import RandomState

rs = RandomState(458794)
n = 1000
n_runs = 100

mses = [None] * n_runs
timestep_targets = [(v+1)/n for v in range(n)]
for i in range(n_runs):
    timestep_rs_1 = rs.uniform(0, 1, n)
    mean_squared_error = MeanSquaredError()
    mses[i] = mean_squared_error(torch.tensor(timestep_rs_1), 
                                 torch.tensor(timestep_targets)) \
        .detach().numpy().item()
print(f"sims' mean MSE: {np.mean(mses):.6f}")
```

MSE of always guessing mean scaled timestep (0.5)

```python
mean_squared_error(torch.tensor(np.array([0.5]).repeat(len(timestep_targets))), 
                   torch.tensor(timestep_targets)) \
    .detach().numpy().item()
```
